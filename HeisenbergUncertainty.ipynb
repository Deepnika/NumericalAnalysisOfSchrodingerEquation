```python
# Imports

import numpy as np
import math
import matplotlib.pyplot as plt
```

#### Assuming exponential wave-function
$\psi (x) = e^{ikx}$ 

Here, the uncertainty in momentum is very small as $p = \hbar k$ (as k is known).

We see that the uncertainty in position of such a system is very large.

```python
i = complex(0.0, 1.0)                                # iota
X = np.linspace(0, 100, 1000)
k0 = 0.5

psi = np.zeros(len(X), complex)

for i in range(len(X)):
    real = np.cos(k0*X[i])
    im = np.sin(k0*X[i])
    psi[i] = complex(real, im)                      # complex wave-function
```

```python
print("", "Exponential Wave-function", "", sep = "--------------------------------------------")

real_psi = np.array(psi.real)
prob_position = abs(psi)**2
ftpsi = np.array(np.fft.fft(psi), dtype = "complex")
prob_momentum = abs(ftpsi)**2

fig, ax = plt.subplots(nrows = 4, figsize = (16, 16))
props = dict(boxstyle='round', facecolor='snow', alpha=1)

# Plot 1
ax[0].plot(X, real_psi, color = "purple")
ax[0].set_ylabel('$Re(\psi)$')
ax[0].set_xlabel('Position')

textstr1 = '\n'.join((r'$\psi (x) = e^{ikx}$',
                      r'$p = \hbar k$'))
ax[0].text(94, 0.95, textstr1, fontsize=14,verticalalignment='top', bbox=props)

# Plot 2
ax[1].plot(X, prob_position, color = 'purple')
ax[1].set_ylabel('P(x)')
ax[1].set_xlabel('Position')

textstr2 = '\n'.join((r'Probability of finding a particle ',
                      r'at a given position (x) = $P(x) = |\psi (x)|^2$',
                      r'$\Delta x$ is large'))
ax[1].text(70, 1.045, textstr2, fontsize=14,verticalalignment='top', bbox=props)

# Plot 3
ax[2].plot(X, ftpsi.real, color = 'purple')
ax[2].set_ylabel('$Re(fft(\psi))$')
ax[2].set_xlabel('Momentum')

textstr3 = '\n'.join((r'Fourier Transform of $\psi (x)$ to visualize',
                      r'the wave in momentum space.'))
ax[2].text(70, 900, textstr3, fontsize=14,verticalalignment='top', bbox=props)

# Plot 4
ax[3].plot(X, prob_momentum, color = 'purple')
ax[3].set_ylabel('$P(p = \hbar k)$')
ax[3].set_xlabel('Momentum')

textstr4 = '\n'.join((r'Probability of finding a particle with a given',
                      r'momentum (p) = $P(p = \hbar k) = |fft(\psi (x))|^2$',
                      r'$\Delta p$ is small'))
ax[3].text(65, 0.9e6, textstr4, fontsize=14,verticalalignment='top', bbox=props)

plt.show()
```

#### Assuming Delta wave-function

Note that the above exponential wave-function (real part that consists of a cosine function) returns a delta function after fourier transform. Conversely, if we take a delta wave-function, the uncertainty in position would be small, but the uncertainty in momentum would be large.

This illustrates two extremes of the Heisenberg Uncertainty principle.
